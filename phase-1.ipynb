### Series: A Pandas like 1d labeled array that can hold any data types

```python
import pandas as pd 
```

```python
data = [100.2, 200, 300]
series = pd.Series(data, index=["a","b","c"])
print(series[series > 150])
```

```python
data1 = {'key': ['K0', 'K1', 'K2', 'K3'],
         'Name':['Jai', 'Princi', 'Gaurav', 'Anuj'], 
        'Age':[27, 24, 22, 32],} 

data2 = {'key': ['K0', 'K1', 'K2', 'K3'],
         'Address':['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'], 
        'Qualification':['Btech', 'B.A', 'Bcom', 'B.hons']} 
 
df = pd.DataFrame(data1)
df1 = pd.DataFrame(data2)  

display(df,df1)

res = pd.merge(df, df1, on='key')
print(res)
```

```python
calories = {"Day 1": 1750, "Day 2": 2100, "Day 3": 1700}

series = pd.Series(calories)

series.loc["Day 3"] += 500
print(series.loc["Day 3"])
print(series)
print(series[series >= 2000])
```

```python
name = {1: 'John', 2: 'Smith', 3: 'Paul', 4: 'Yuvraj', 5: 'Singh', 6: 'Kohli'}

series = pd.Series(name)

print(series)
```

### Data Frames: A tabular data structure with rows and columns. (2 Dimensional) Similar to Excel Spreadsheet

```python
data = {"Name": ['Jai', 'Princi', 'Gaurav', 'Anuj'],
        "Age": [27, 24, 22, 32]}

df = pd.DataFrame(data, index=["Emp1", "Emp2", "Emp3", "Emp4"])
# print(df)
# print(df.loc["Emp2"])
# print(df.iloc[1])

#Adding a new column
df["Job"] = ["Engineer", "Doctor", "Artist", "Lawyer"]
# print(df)

#Adding new row
new_rows = pd.DataFrame([{"Name": "Karan", "Age": 29, "Job": "Architect"},{"Name": "Soni", "Age": 45, "Job": "Mech"}], index=["Emp5", "Emp6"])
df = pd.concat([df, new_rows])
print(df)
```
